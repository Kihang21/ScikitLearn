```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('KMeansData.csv')
dataset[:5]
```

```python
X = dataset.iloc[:, :].values # K-Means는 비지도 학습이기 때문에 종속변수 없음, Y 필요없음
# X = dataset.values
# X = dataset.to_numpy() # 공식 홈페이지 권장
X[:5]
```

### 데이터 시각화 (전체 데이터 분포 확인)

```python
plt.scatter(X[:, 0], X[:, 1]) # x축 : hour, y축 : score
plt.title('Score by hours')
plt.xlabel('houes')
plt.ylabel('score')
plt.show()
```

### 데이터 시각화 (축 범위 동일)

```python
plt.scatter(X[:, 0], X[:, 1]) # x축 : hour, y축 : score
plt.title('Score by hours')
plt.xlabel('houes')
plt.xlim(0, 100)
plt.ylabel('score')
plt.ylim(0, 100)
plt.show()
```

### 피쳐 스케일링 (Feature Scaling)

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X[:5]
```

### 데이터 시각화 (스케일링된 데이터)

```python
plt.figure(figsize=(5, 5)) # 그래프 정사각형 만들기
plt.scatter(X[:, 0], X[:, 1]) # x축 : hour, y축 : score
plt.title('Score by hours')
plt.xlabel('houes')
plt.ylabel('score')
plt.show()
```

### 최적의 K 값 찾기 (엘보우 방식 Elbow Method)

```python
from sklearn.cluster import KMeans
inertia_list = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    # 일반 kmeans 방식의 랜덤으로 중심점을 잡으면 문제가 될수 있기 때문에 kmeans++를 사용함
    kmeans.fit(X)
    inertia_list.append(kmeans.inertia_) # 각 지점으로부터의 클러스터 중심(centroid) 까지의 거리의 제곱의 합
    
plt.plot(range(1, 11), inertia_list)
plt.title('Elbow Method')
plt.xlabel('n_clusters')
plt.ylabel('inertia')
plt.show()
```

### 최적의 K (4) 값 KMeans 학습

```python
K = 4 # 최적의 K 값
```

```python
kmeans = KMeans(n_clusters=K, random_state=0)
# kmeans.fit(X)
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

### 데이터 시각화 (최적의 K)

```python
centers = kmeans.cluster_centers_ # 클러스터의 중심점 (centroid) 좌표
centers
```

```python
for cluster in range(K):
    plt.scatter(X[y_kmeans == cluster, 0], X[y_kmeans == cluster, 1], s=100, edgecolor='black') # 각 데이터
    plt.scatter(centers[cluster, 0], centers[cluster, 1], s=300, edgecolor='black', color='yellow', marker='s') # 중심점 네모
    plt.text(centers[cluster, 0], centers[cluster, 1], cluster, va='center', ha='center') # 클러스터 텍스트 출력
    
plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

### 데이터 시각화 (스케일링 원복)

```python
X_org = sc.inverse_transform(X) # Feature Scaling 된 데이터를 다시 원복
X_org[:5]
```

```python
centers_org = sc.inverse_transform(centers) # Feature Scaling 된 데이터를 다시 원복
centers_org
```

```python
for cluster in range(K):
    plt.scatter(X_org[y_kmeans == cluster, 0], X_org[y_kmeans == cluster, 1], s=100, edgecolor='black') # 각 데이터
    plt.scatter(centers_org[cluster, 0], centers_org[cluster, 1], s=300, edgecolor='black', color='yellow', marker='s') # 중심점 네모
    plt.text(centers_org[cluster, 0], centers_org[cluster, 1], cluster, va='center', ha='center') # 클러스터 텍스트 출력
    
plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```
